```
nodupes_df = df.drop_duplicates(subset=['artifact_id', keep='last')
```

```
ordered_df = nodupes_df[['artifact_id','museum_no','designation','publications_key','provenience','period','collections','publications_exact_ref','height','thickness','width','genres','artifact_preservation','cdli_comments','provenience_comments','seal_information']]
```

```
column_target = 'provenience'
unique = len(pd.unique(ordered_df[column_target]))
unique_2 = len(ordered_df[column_target].unique())
print(f"Number of unique values in '{column_target}': {unique}")
```

```
locations = pd.unique(ordered_df['provenience'])
print(locations)
```

```
# counting the number of tablet based the location

#Susa
target_location_susa = "Susa (mod. Shush)"
filtered_df_susa = ordered_df.loc[ordered_df['provenience'] == target_location_susa]


#Anshan
target_location_anshan = "Anšan (mod. Tell Malyan)"
filtered_df_Anshan = ordered_df.loc[ordered_df['provenience'] == target_location_anshan]

#Yahye
target_location_yahya = "uncertain (mod. Tepe Yahya)"
filtered_df_yahya = ordered_df.loc[ordered_df['provenience'] == target_location_yahya]

```

```
total_tablets = len(filtered_df_susa)

print(f"Number of tablets found in {target_location_susa}:", total_tablets)
```

```
total_tablets = len(filtered_df_Anshan)

print(f"Number of tablets found in {target_location_anshan}:", total_tablets)
```

```
total_tablets = len(filtered_df_yahya)

print(f"Number of tablets found in {target_location_yahya}:", total_tablets)
```
